### 4.0  If-Statements 

##### 4.1  Define a function  : If-Statements with User-Defined Functions

```python
# define a statement with if usung user defined functions
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: 
    return 'Not enough data'
```

```python
# Apply the function
#the warning provide for alternative method using loc 
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df.head(2)
```

```python
# value count of the create column
df['price_range'].value_counts(dropna = False)
```

##### 4. 2 If-Statements with the loc() Function

```python
#### If-Statements with loc function for high range product
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# If-Statements with loc function for mid_range product
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
# If-Statements with loc function for low range product 
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df.head(2)
```

```python
# frequency of the price range 
df['price_range_loc'].value_counts(dropna = False)
```
