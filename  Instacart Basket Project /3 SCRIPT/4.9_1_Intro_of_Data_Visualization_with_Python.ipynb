### 3.0 Data wrangling

```python
# Getting the columns 
df_customers.columns
```

```python
### Renaming of the specific columns
df_customers.rename(columns = {'Surnam':'Last Name', 'STATE':'State', 'n_dependants': 'No_of_dependants', 'fam_status'
                    :'family_status'}, inplace = True)
```

```python
df_customers.head()
```

```python
#checking for the mix datatype in the dataframe
for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers[weird]) > 0:
    print (col)
```

```python
# checking the total number of rows  of the missing column 
df_customers.isnull().sum()
```

### 4.0 Creating a subset 

```python
#4.1 Creating a subset of  the nan using isnull(== True)
df_cus_nan = df_customers[df_customers['First Name']. isnull() == True]
```

```python
df_cus_nan.head(3)
```

```python
#checking the shape of the subset wuth NAN
df_cus_nan.shape
```

```python
# Creating a subset of  the filter dataframe without NAN using isnull(== False)
df_cus_clean = df_customers[df_customers['First Name']. isnull() == False]
```

```python
df_cus_clean.head(5)
```

```python
# Checking for mix data type in the [df_cus_clean]
for col in df_cus_clean.columns.tolist():
  weird = (df_cus_clean[[col]].applymap(type) != df_cus_clean[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cus_clean[weird]) > 0:
    print (col)
```

```python
# checking the number of  rows and column of the new subset[df_cus_clean.shape]
df_cus_clean.shape
```

### 5.0 Checking for duplicate 

```python
# checking for duplicate in the df_cus_clean
dup_cus_clean = df_cus_clean[df_cus_clean.duplicated()]
```

```python
dup_cus_clean
```

###### No duplicate was found in the dataset

```python
# to be sure, i use duplicate function to further eliminate any form of duplicate 
df_cus = df_cus_clean.drop_duplicates()
```

```python
df_cus.head(5)
```

```python
#checking the rows and columns of the clean dataset
df_cus.shape
```
